```python
from langchain_community.llms import CTransformers
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
llm = CTransformers(model= "TheBloke/Llama-2-7B-Chat-GGML",
                    model_type='llama',
                    config = {"seed":42, "temperature": 0.6, "max_new_tokens": 100,"repetition_penalty": 1.1})
```

```python
prompt = """You are an expert in HR domain. What are the respnsibility of {title}. Response should be in bullet points."""
prompt_template = PromptTemplate(input_variables = ['title'], template= prompt)
```

```python
chain = LLMChain(prompt = prompt_template, llm = llm)
```

```python
response = chain({'title': 'Java Developer'})
```
